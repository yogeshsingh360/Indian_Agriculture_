```python
import seaborn as sns
import matplotlib.pyplot as plt
numerical_columns = ['area','temperature_2m_mean','precipitation_sum','relative_humidity_2m_mean','wind_speed_10m_mean','latitude','longitude']

for col in numerical_columns:  # Replace with your actual column names
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot for {col}")
    plt.show()
```

```python
# Graphical analysis of crop_yield data with respect to independent features
for i in df.columns.drop(['season','crop','crop_yield']):
    plt.scatter(df[i],df['crop_yield'])
    plt.xlabel(i)
    plt.ylabel('crop_yield')
    plt.show()
```

```python
# checking mini and max values of crop_yield and area
max(df['crop_yield']),min(df['crop_yield']),max(df['area']),min(df['area'])
```

```python
# from graph the yield and area values have very large value differences. So we apply Log transformation on crop_yield and area column
df['crop_yield']=np.log1p(df['crop_yield'])
df['area']=np.log1p(df['area'])
```

```python
# Data analysis after applying logp1
for i in df.columns.drop(['crop_yield']):
    plt.scatter(df[i],df['crop_yield'])
    plt.xlabel(i)
    plt.ylabel('crop_yield')
    plt.show()
```

```python
df = df[~df['crop_yield'].isin(['inf', np.inf])]
```

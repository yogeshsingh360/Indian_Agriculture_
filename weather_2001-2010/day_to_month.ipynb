```python
import pandas as pd 
import numpy as np 
```

```python
# for first 300+ districts
# converting day by day weather_data into month wise by taking mean
data=[]
for i in range(0,12):
    if i<10:
        df=pd.read_csv(f'historical_weather_300_data200{i}.csv')
    else:
        df=pd.read_csv(f'historical_weather_300_data20{i}.csv')
    df['date'] = pd.to_datetime(df['date'])  # Convert to datetime format
    
    # Add year and month columns
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    
    # Group by state, district, year, and month, and aggregate numerical columns
    monthly_df = df.groupby(['state', 'district', 'year', 'month']).agg({
        'precipitation_sum': 'mean',
        'relative_humidity_2m_mean': 'mean',  # or 'sum' if needed
        'latitude': 'first',
        'longitude': 'first', 
        'temperature_2m_mean':'mean', 
        'wind_speed_10m_mean':'mean'
    }).reset_index()
    data.append(monthly_df)
    
    
```

```python
len(data)
```

```python
# for next 300+ districts
# converting day by day weather_data into month wise by taking mean
data_1=[]
for i in range(0,12):
    if i<10:
        df=pd.read_csv(f'historical_weather_600_data200{i}.csv')
    else:
        df=pd.read_csv(f'historical_weather_600_data20{i}.csv')
    df['date'] = pd.to_datetime(df['date'])  # Convert to datetime format
    
    # Add year and month columns
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    
    # Group by state, district, year, and month, and aggregate numerical columns
    monthly_df = df.groupby(['state', 'district', 'year', 'month']).agg({
        'precipitation_sum': 'mean',
        'relative_humidity_2m_mean': 'mean',  # or 'sum' if needed
        'latitude': 'first',
        'longitude': 'first', 
        'temperature_2m_mean':'mean', 
        'wind_speed_10m_mean':'mean'
    }).reset_index()
    data_1.append(monthly_df)
```

```python
len(data_1)
```

```python
# data for each year saved in data list now coverting it into dataframe
combined_df_1 = pd.concat(data, ignore_index=True)
```

```python
combined_df_1
```

```python
# data for each year saved in data list now coverting it into dataframe
combined_df_2 = pd.concat(data_1, ignore_index=True)
```

```python
combined_df_2
```

```python
# combined data of all 600+ districts
f = pd.concat([combined_df_1,combined_df_2], ignore_index=True)
```

```python
# saving final data in csv
f.to_csv('final_weather.csv',index=False)
```
